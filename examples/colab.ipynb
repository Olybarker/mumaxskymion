# Running mumax+

Now it's up to you! The example below solves Standard Problem 4 using mumax⁺, but feel free to create a copy of this notebook to solve another micromagnetic problem of your choice.

```python
# =========================================================
# CoFeB-like skyrmion relaxation across grain/disorder cases
# mumax+ / Google Colab script
#
# Cases:
#   1. no_grains
#   2. ku1_only
#   3. msat_only
#   4. dmi_only
#   5. all_three
# =========================================================


import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import zipfile
import pandas as pd
import gc
try:
    from google.colab import files
    IN_COLAB = True
except Exception:
    IN_COLAB = False
from mumaxplus import Ferromagnet, Grid, World
from mumaxplus.util import VoronoiTessellator

# =========================================================
# 0) Input design mask
# =========================================================
# Upload your design mask to Colab as:

#
# Default convention:
#   black pixels = magnetic material
#   white pixels = non-magnetic background

shape_path = "/content/full_mask.png"

if not os.path.exists(shape_path):
    raise FileNotFoundError(
        f"Could not find {shape_path}. Upload full_mask.png to /content or change shape_path."
    )

print("Using shape file:", shape_path)

# =========================================================
# 1) Grid and cell size
# =========================================================
USE_IMAGE_SIZE = True

if USE_IMAGE_SIZE:
    img_tmp = Image.open(shape_path).convert("L")
    Nx, Ny = img_tmp.size
    img_tmp.close()
else:
    Nx = 512
    Ny = 512

Nz = 10

# For your reduced design/current mesh, 2 nm is the natural starting point.
dx = 2e-9
dy = 2e-9

# Set this to your real CoFeB magnetic thickness.
# 1.0 nm is a practical starting value.
dz = 1e-10

print(f"Grid: Nx={Nx}, Ny={Ny}, Nz={Nz}")
print(f"Cell size: dx={dx*1e9:.2f} nm, dy={dy*1e9:.2f} nm, dz={dz*1e9:.2f} nm")

# =========================================================
# 2) CoFeB-like material parameters
# =========================================================
MS0 = 1100e3       # A/m
AEX0 = 15e-12      # J/m
D0 = 1.2e-3        # J/m^2
KU10 = 1.2e6       # J/m^3

# High damping is fine for relaxation.
ALPHA = 0.5

# Static out-of-plane field during relaxation.
# For a skyrmion with core down and background up:
#   BIAS_BZ_T > 0 tends to shrink the core
#   BIAS_BZ_T < 0 tends to expand the core
BIAS_BZ_T = 40e-3

# =========================================================
# 3) Grain/disorder settings
# =========================================================
# CoFeB-like effective disorder patch size.
# For mumax+, 20-30 nm is a reasonable starting range.
grain_size = 50e-9

# mumax+ can handle many region IDs.
MAX_REGION_ID = 1500

# Disorder amplitudes.
# Ku is usually the main disorder channel for PMA CoFeB.
KU1_SIGMA = 0.05
MS_SIGMA  = 0.05
D_SIGMA   = 0.05

# More physical for interface-driven PMA/DMI disorder.
# Same random variable controls Ku and DMI.
CORRELATE_KU_DMI = True

RANDOM_SEED = 12345

# Mask convention.
# False: black pixels are magnetic.
# True: white pixels are magnetic.
INVERT_MASK = False

# =========================================================
# 4) Skyrmion seed settings
# =========================================================
SK_RADIUS = 100e-9
SK_WALL_WIDTH = 8e-9

# -1 means core down, background up.
# +1 means core up, background down.
SK_CORE_POLARITY = -1

# Néel chirality.
# Try +1 first. If everything collapses, test -1.
NEEL_CHIRALITY = +1

# Optional manual skyrmion center.
# Use None to choose the magnetic cell nearest the image center.
MANUAL_CENTER_IX_IY = None
# Example:
# MANUAL_CENTER_IX_IY = (256, 256)

# =========================================================
# 5) Relaxation controls
# =========================================================
RELAX_METHOD = "relax"       # "minimize" or "relax"

MINIMIZE_TOL = 1e-5
MINIMIZE_NSAMPLES = 10

RELAX_TOL = 1e-5

SAVE_INITIAL_OVF = True
SAVE_FINAL_OVF = True
SAVE_MATERIAL_OVFS = False

# =========================================================
# Choose which variation to run
# =========================================================
# Options:
#   "no_grains"
#   "ku1_only"
#   "msat_only"
#   "dmi_only"
#   "all_three"
#   "all"
#
# Change only this line when you want to run a different variation.
RUN_VARIATION = "all_three"
VALID_CASES = [
    "no_grains",
    "ku1_only",
    "msat_only",
    "dmi_only",
    "all_three",
]

if RUN_VARIATION == "all":
    cases = VALID_CASES
elif RUN_VARIATION in VALID_CASES:
    cases = [RUN_VARIATION]
else:
    raise ValueError(f"Unknown RUN_VARIATION: {RUN_VARIATION}")

print("Cases to run:", cases)

# =========================================================
# 6) Output folders
# =========================================================
base_out_dir = "cofeb_grain_skyrmion_relax_outputs"

plot_dir = os.path.join(base_out_dir, "plots")
ovf_dir = os.path.join(base_out_dir, "ovf")
csv_dir = os.path.join(base_out_dir, "csv")
grain_dir = os.path.join(base_out_dir, "grain_maps")

for d in [base_out_dir, plot_dir, ovf_dir, csv_dir, grain_dir]:
    os.makedirs(d, exist_ok=True)

# =========================================================
# 7) Load mask geometry
# =========================================================
def load_geometry_from_image(path, nx, ny, nz, threshold=128, invert=False):
    img = Image.open(path).convert("L")
    img = img.resize((nx, ny), Image.Resampling.NEAREST)
    arr = np.array(img)

    if invert:
        mask2d = arr > threshold
    else:
        mask2d = arr < threshold

    geometry = np.repeat(mask2d[np.newaxis, :, :], nz, axis=0).astype(bool)

    magnetic_fraction = geometry.mean()
    print(f"Magnetic fraction of grid: {magnetic_fraction:.4f}")

    if magnetic_fraction < 0.001:
        print("WARNING: geometry is almost empty. Try INVERT_MASK = True.")
    if magnetic_fraction > 0.999:
        print("WARNING: geometry is almost full. Check whether the mask is inverted.")

    return geometry

geometry = load_geometry_from_image(
    shape_path,
    Nx,
    Ny,
    Nz,
    invert=INVERT_MASK,
)

geometry2d = geometry[0]

plt.figure(figsize=(6, 6))
plt.imshow(geometry2d, cmap="gray", origin="lower")
plt.title("Imported geometry: white = magnetic")
plt.axis("off")
plt.tight_layout()
plt.savefig(os.path.join(plot_dir, "imported_geometry.png"), dpi=200, bbox_inches="tight")
plt.show()

# =========================================================
# 8) Generate one shared Voronoi grain map
# =========================================================
region_world = World(cellsize=(dx, dy, dz))
region_grid = Grid((Nx, Ny, Nz))

tessellator = VoronoiTessellator(
    grain_size,
    seed=RANDOM_SEED,
    max_idx=MAX_REGION_ID,
)

regions = tessellator.generate(region_world, region_grid, make_2d=True)
regions = np.asarray(regions).astype(np.int32)

if regions.ndim == 2:
    regions = regions[np.newaxis, :, :]

if regions.shape[0] == 1 and Nz > 1:
    regions = np.repeat(regions, Nz, axis=0)

if regions.shape != geometry.shape:
    raise ValueError(
        f"regions shape {regions.shape} does not match geometry shape {geometry.shape}"
    )

regions2d = regions[0]

active_region_ids = np.unique(regions[geometry])
active_region_ids = [int(i) for i in active_region_ids]

print("Total Voronoi region IDs in full box:", len(np.unique(regions)))
print("Active grains inside magnetic geometry:", len(active_region_ids))
print("Lowest active region ID:", min(active_region_ids))
print("Highest active region ID:", max(active_region_ids))

# Save region map.
region_map = np.full((Ny, Nx), np.nan, dtype=np.float32)
region_map[geometry2d] = regions2d[geometry2d]

cmap_regions = plt.get_cmap("nipy_spectral").copy()
cmap_regions.set_bad(color="white")

plt.figure(figsize=(6, 6))
plt.imshow(region_map, origin="lower", cmap=cmap_regions, interpolation="nearest")
plt.title(f"Voronoi grains, grain size = {grain_size*1e9:.1f} nm")
plt.axis("off")
plt.tight_layout()
plt.savefig(os.path.join(grain_dir, "voronoi_regions.png"), dpi=200, bbox_inches="tight")
plt.show()

# =========================================================
# 9) Pre-generate shared disorder values
# =========================================================
rng = np.random.default_rng(RANDOM_SEED)

ku1_by_region = {}
msat_by_region = {}
dmi_by_region = {}

for rid in active_region_ids:
    if CORRELATE_KU_DMI:
        eta_interface = rng.normal()
        eta_ms = rng.normal()

        ku1_by_region[rid] = max(0.0, KU10 * (1.0 + KU1_SIGMA * eta_interface))
        dmi_by_region[rid] = max(0.0, D0   * (1.0 + D_SIGMA   * eta_interface))
        msat_by_region[rid] = max(1.0, MS0 * (1.0 + MS_SIGMA  * eta_ms))
    else:
        ku1_by_region[rid] = max(0.0, KU10 * (1.0 + KU1_SIGMA * rng.normal()))
        msat_by_region[rid] = max(1.0, MS0 * (1.0 + MS_SIGMA  * rng.normal()))
        dmi_by_region[rid] = max(0.0, D0   * (1.0 + D_SIGMA   * rng.normal()))

# =========================================================
# 10) Parameter maps and plotting helpers
# =========================================================
def make_param_map(default_value, values_by_region=None):
    data = np.full((Ny, Nx), np.nan, dtype=np.float32)
    data[geometry2d] = default_value

    if values_by_region is not None:
        for rid, value in values_by_region.items():
            data[(regions2d == rid) & geometry2d] = value

    return data

ku1_uniform_map = make_param_map(KU10)
msat_uniform_map = make_param_map(MS0)
dmi_uniform_map = make_param_map(D0)

ku1_var_map = make_param_map(KU10, ku1_by_region)
msat_var_map = make_param_map(MS0, msat_by_region)
dmi_var_map = make_param_map(D0, dmi_by_region)

def cmap_with_white_background(name):
    cmap = plt.get_cmap(name).copy()
    cmap.set_bad(color="white")
    return cmap

def plot_map(ax, data, title, cmap_name, units_label=None):
    cmap = cmap_with_white_background(cmap_name)
    im = ax.imshow(data, cmap=cmap, origin="lower", interpolation="nearest")
    ax.set_title(title)
    ax.axis("off")

    cb = plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    if units_label is not None:
        cb.set_label(units_label)

def save_global_disorder_panel():
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    axs = axs.ravel()

    plot_map(axs[0], region_map, "Voronoi regions", "nipy_spectral")
    plot_map(axs[1], ku1_var_map, "Ku1 disorder", "viridis", "J/m³")
    plot_map(axs[2], msat_var_map, "Msat disorder", "magma", "A/m")
    plot_map(axs[3], dmi_var_map, "DMI disorder", "plasma", "J/m²")

    plt.tight_layout()
    path = os.path.join(grain_dir, "shared_disorder_parameter_maps.png")
    plt.savefig(path, dpi=200, bbox_inches="tight")
    plt.show()
    print("Saved:", path)

save_global_disorder_panel()

# =========================================================
# 11) Save shared grain-value CSV
# =========================================================
def save_grain_value_csv():
    rows = []

    for rid in active_region_ids:
        rows.append({
            "region": rid,
            "ku1_J_per_m3": ku1_by_region[rid],
            "msat_A_per_m": msat_by_region[rid],
            "dmi_J_per_m2": dmi_by_region[rid],
        })

    df = pd.DataFrame(rows)
    path = os.path.join(csv_dir, "shared_grain_disorder_values.csv")
    df.to_csv(path, index=False)
    print("Saved:", path)

save_grain_value_csv()

# =========================================================
# 12) Skyrmion seed construction
# =========================================================
def choose_center_cell(mask2d):
    cy = mask2d.shape[0] // 2
    cx = mask2d.shape[1] // 2

    if mask2d[cy, cx]:
        return cx, cy

    ys, xs = np.where(mask2d)
    if len(xs) == 0:
        raise ValueError("No magnetic cells found in geometry.")

    dist2 = (xs - cx)**2 + (ys - cy)**2
    idx = np.argmin(dist2)

    return int(xs[idx]), int(ys[idx])

if MANUAL_CENTER_IX_IY is None:
    center_ix, center_iy = choose_center_cell(geometry2d)
else:
    center_ix, center_iy = MANUAL_CENTER_IX_IY
    if not geometry2d[center_iy, center_ix]:
        raise ValueError("Manual skyrmion center is outside magnetic geometry.")

print(f"Skyrmion center cell: ix={center_ix}, iy={center_iy}")
print(f"Skyrmion center position: x={center_ix*dx*1e9:.1f} nm, y={center_iy*dy*1e9:.1f} nm")

def make_neel_skyrmion_seed(
    nx,
    ny,
    nz,
    dx,
    dy,
    center_ix,
    center_iy,
    radius,
    wall_width,
    core_polarity=-1,
    chirality=+1,
):
    yy, xx = np.meshgrid(np.arange(ny), np.arange(nx), indexing="ij")

    x = (xx - center_ix) * dx
    y = (yy - center_iy) * dy

    r = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)

    # mz goes from core_polarity at center to -core_polarity far away.
    mz = -core_polarity * np.tanh((r - radius) / wall_width)

    # In-plane component peaks at the domain wall.
    mrad = 1.0 / np.cosh((r - radius) / wall_width)

    # Néel radial texture.
    mx = chirality * mrad * np.cos(phi)
    my = chirality * mrad * np.sin(phi)

    mx *= -core_polarity
    my *= -core_polarity

    # Normalize.
    norm = np.sqrt(mx**2 + my**2 + mz**2)
    mx /= norm
    my /= norm
    mz /= norm

    m = np.zeros((3, nz, ny, nx), dtype=np.float32)

    for iz in range(nz):
        m[0, iz, :, :] = mx
        m[1, iz, :, :] = my
        m[2, iz, :, :] = mz

    return m

m0 = make_neel_skyrmion_seed(
    Nx,
    Ny,
    Nz,
    dx,
    dy,
    center_ix,
    center_iy,
    SK_RADIUS,
    SK_WALL_WIDTH,
    core_polarity=SK_CORE_POLARITY,
    chirality=NEEL_CHIRALITY,
)

# Set non-magnetic cells to background for cleaner plots/OVFs.
background_mz = -SK_CORE_POLARITY

m0[0, :, ~geometry2d] = 0.0
m0[1, :, ~geometry2d] = 0.0
m0[2, :, ~geometry2d] = background_mz

initial_mz_plot = np.full((Ny, Nx), np.nan, dtype=np.float32)
initial_mz_plot[geometry2d] = m0[2, 0, :, :][geometry2d]

plt.figure(figsize=(6, 6))
plt.imshow(initial_mz_plot, origin="lower", cmap="RdBu_r", vmin=-1, vmax=1)
plt.colorbar(label="mz")
plt.title("Initial skyrmion seed")
plt.axis("off")
plt.tight_layout()
plt.savefig(os.path.join(plot_dir, "initial_skyrmion_seed_mz.png"), dpi=200, bbox_inches="tight")
plt.show()

# =========================================================
# 13) Build magnet for each disorder case
# =========================================================
def build_magnet(case_name):
    case_world = World(cellsize=(dx, dy, dz))
    case_grid = Grid((Nx, Ny, Nz))

    if case_name == "no_grains":
        magnet = Ferromagnet(
            case_world,
            case_grid,
            geometry=geometry,
            name=case_name,
        )
    else:
        magnet = Ferromagnet(
            case_world,
            case_grid,
            geometry=geometry,
            regions=regions,
            name=case_name,
        )

    magnet.alpha = ALPHA
    magnet.aex = AEX0
    magnet.msat = MS0
    magnet.ku1 = KU10
    magnet.anisU = (0, 0, 1)
    magnet.dmi_tensor.set_interfacial_dmi(D0)

    magnet.bias_magnetic_field = (0, 0, BIAS_BZ_T)

    # No current torques for relaxation-only runs.
    magnet.enable_zhang_li_torque = False
    magnet.enable_slonczewski_torque = False

    # Initial skyrmion state.
    magnet.magnetization = m0

    # no_grains: uniform material, no region map, no disorder.
    if case_name == "no_grains":
        return magnet

    # Grain-by-grain disorder.
    for rid in active_region_ids:
        if case_name in ["ku1_only", "all_three"]:
            magnet.ku1.set_in_region(rid, ku1_by_region[rid])

        if case_name in ["msat_only", "all_three"]:
            magnet.msat.set_in_region(rid, msat_by_region[rid])

        if case_name in ["dmi_only", "all_three"]:
            d_i = dmi_by_region[rid]
            magnet.dmi_tensor.xxz.set_in_region(rid, d_i)
            magnet.dmi_tensor.yyz.set_in_region(rid, d_i)

    return magnet

# =========================================================
# 14) Analysis helpers
# =========================================================
def get_m_array(magnet):
    m = np.asarray(magnet.magnetization.eval())

    expected_shape = (3, Nz, Ny, Nx)
    if m.shape != expected_shape:
        raise ValueError(f"Unexpected magnetization shape: got {m.shape}, expected {expected_shape}")

    return m

def estimate_topological_charge(m_array, mask2d, dx, dy):
    """
    Approximate finite-difference topological charge:
        Q = 1/(4*pi) integral m · (dm/dx cross dm/dy) dx dy

    For irregular masked shapes this is approximate, but useful for checking
    whether the relaxed state remains skyrmion-like.
    """
    m = m_array[:, 0, :, :]  # shape: 3, Ny, Nx

    dmdy, dmdx = np.gradient(m, dy, dx, axis=(1, 2))

    m_vec = np.moveaxis(m, 0, -1)
    dmdx_vec = np.moveaxis(dmdx, 0, -1)
    dmdy_vec = np.moveaxis(dmdy, 0, -1)

    cross = np.cross(dmdx_vec, dmdy_vec)
    q_density = np.einsum("ijk,ijk->ij", m_vec, cross)

    q_density = np.where(mask2d, q_density, 0.0)

    Q = np.sum(q_density) * dx * dy / (4 * np.pi)

    return float(Q)

def estimate_reversed_core_area_radius(m_array, mask2d, core_polarity):
    mz = m_array[2, 0, :, :]

    if core_polarity == -1:
        core_mask = (mz < 0.0) & mask2d
    else:
        core_mask = (mz > 0.0) & mask2d

    area = np.sum(core_mask) * dx * dy
    radius = np.sqrt(area / np.pi) if area > 0 else 0.0

    return float(area), float(radius)

def save_case_parameter_panel(case_name):
    case_plot_dir = os.path.join(plot_dir, case_name)
    os.makedirs(case_plot_dir, exist_ok=True)

    if case_name == "no_grains":
        fig, axs = plt.subplots(1, 3, figsize=(18, 5))
        plot_map(axs[0], ku1_uniform_map, "Ku1 uniform, no grains", "viridis", "J/m³")
        plot_map(axs[1], msat_uniform_map, "Msat uniform, no grains", "magma", "A/m")
        plot_map(axs[2], dmi_uniform_map, "DMI uniform, no grains", "plasma", "J/m²")

    elif case_name == "ku1_only":
        fig, axs = plt.subplots(1, 2, figsize=(12, 5))
        plot_map(axs[0], region_map, "Voronoi regions", "nipy_spectral")
        plot_map(axs[1], ku1_var_map, "Ku1 varied", "viridis", "J/m³")

    elif case_name == "msat_only":
        fig, axs = plt.subplots(1, 2, figsize=(12, 5))
        plot_map(axs[0], region_map, "Voronoi regions", "nipy_spectral")
        plot_map(axs[1], msat_var_map, "Msat varied", "magma", "A/m")

    elif case_name == "dmi_only":
        fig, axs = plt.subplots(1, 2, figsize=(12, 5))
        plot_map(axs[0], region_map, "Voronoi regions", "nipy_spectral")
        plot_map(axs[1], dmi_var_map, "DMI varied", "plasma", "J/m²")

    elif case_name == "all_three":
        fig, axs = plt.subplots(2, 2, figsize=(12, 10))
        axs = axs.ravel()
        plot_map(axs[0], region_map, "Voronoi regions", "nipy_spectral")
        plot_map(axs[1], ku1_var_map, "Ku1 varied", "viridis", "J/m³")
        plot_map(axs[2], msat_var_map, "Msat varied", "magma", "A/m")
        plot_map(axs[3], dmi_var_map, "DMI varied", "plasma", "J/m²")

    else:
        raise ValueError(f"Unknown case: {case_name}")

    fig.suptitle(case_name, fontsize=16)
    plt.tight_layout()

    path = os.path.join(case_plot_dir, f"{case_name}_parameter_panel.png")
    plt.savefig(path, dpi=200, bbox_inches="tight")
    plt.show()
    print("Saved:", path)

def plot_case_results(case_name, m_final):
    case_plot_dir = os.path.join(plot_dir, case_name)
    os.makedirs(case_plot_dir, exist_ok=True)

    # Relaxed mz.
    mz_data = np.full((Ny, Nx), np.nan, dtype=np.float32)
    mz_data[geometry2d] = m_final[2, 0, :, :][geometry2d]

    plt.figure(figsize=(6, 6))
    plt.imshow(mz_data, origin="lower", cmap="RdBu_r", vmin=-1, vmax=1)
    plt.colorbar(label="mz")
    plt.title(f"{case_name}: relaxed mz")
    plt.axis("off")
    plt.tight_layout()

    path = os.path.join(case_plot_dir, f"{case_name}_relaxed_mz.png")
    plt.savefig(path, dpi=200, bbox_inches="tight")
    plt.show()
    print("Saved:", path)

    # mx, my, mz panel.
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    component_names = ["mx", "my", "mz"]

    for c, ax in enumerate(axs):
        data = np.full((Ny, Nx), np.nan, dtype=np.float32)
        data[geometry2d] = m_final[c, 0, :, :][geometry2d]

        im = ax.imshow(data, origin="lower", cmap="RdBu_r", vmin=-1, vmax=1)
        ax.set_title(f"{case_name}: {component_names[c]}")
        ax.axis("off")
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

    plt.tight_layout()

    path = os.path.join(case_plot_dir, f"{case_name}_relaxed_components.png")
    plt.savefig(path, dpi=200, bbox_inches="tight")
    plt.show()
    print("Saved:", path)

def save_material_ovfs(magnet, case_name, case_ovf_dir):
    if not SAVE_MATERIAL_OVFS:
        return

    if case_name in ["ku1_only", "all_three"]:
        magnet.ku1.save_ovf(os.path.join(case_ovf_dir, f"{case_name}_ku1.ovf"))

    if case_name in ["msat_only", "all_three"]:
        magnet.msat.save_ovf(os.path.join(case_ovf_dir, f"{case_name}_msat.ovf"))

    if case_name in ["dmi_only", "all_three"]:
        magnet.dmi_tensor.xxz.save_ovf(os.path.join(case_ovf_dir, f"{case_name}_dmi_xxz.ovf"))
        magnet.dmi_tensor.yyz.save_ovf(os.path.join(case_ovf_dir, f"{case_name}_dmi_yyz.ovf"))


def zip_and_download_finished_case(case_name):
    """
    Zip outputs for the completed variation and download immediately.
    This does not change the simulation physics or parameter values.
    """
    case_zip_name = f"{case_name}_finished_outputs.zip"
    case_zip_path = os.path.join(base_out_dir, case_zip_name)

    case_plot_dir = os.path.join(plot_dir, case_name)
    case_ovf_dir = os.path.join(ovf_dir, case_name)

    include_dirs = [
        case_plot_dir,
        case_ovf_dir,
        csv_dir,
        grain_dir,
    ]

    with zipfile.ZipFile(case_zip_path, "w", compression=zipfile.ZIP_DEFLATED) as z:
        for folder in include_dirs:
            if not os.path.exists(folder):
                continue

            for root, dirs, files_here in os.walk(folder):
                for f in files_here:
                    if f.endswith((".png", ".csv", ".ovf", ".txt")):
                        path = os.path.join(root, f)
                        arcname = os.path.relpath(path, base_out_dir)
                        z.write(path, arcname)

    print("Created:", case_zip_path)

    if IN_COLAB:
        print(f"Downloading completed variation: {case_name}")
        files.download(case_zip_path)
    else:
        print("Not in Colab, so automatic download skipped.")

    return case_zip_path

# =========================================================
# 15) Initial skyrmion diagnostics
# =========================================================
Q_initial = estimate_topological_charge(m0, geometry2d, dx, dy)

area_initial, radius_initial = estimate_reversed_core_area_radius(
    m0,
    geometry2d,
    SK_CORE_POLARITY,
)

print(f"Initial approximate Q: {Q_initial:+.3f}")
print(f"Initial reversed-core radius: {radius_initial*1e9:.2f} nm")

# =========================================================
# 16) Relax skyrmion in each case
# =========================================================
summary_rows = []

for case_name in cases:
    print("\n" + "=" * 72)
    print("Running skyrmion relaxation case:", case_name)
    print("=" * 72)

    case_ovf_dir = os.path.join(ovf_dir, case_name)
    os.makedirs(case_ovf_dir, exist_ok=True)

    save_case_parameter_panel(case_name)

    magnet = build_magnet(case_name)

    print("Initial average m:", magnet.magnetization.average())

    if SAVE_INITIAL_OVF:
        magnet.magnetization.save_ovf(
            os.path.join(case_ovf_dir, f"{case_name}_initial_magnetization.ovf")
        )

    save_material_ovfs(magnet, case_name, case_ovf_dir)

    if RELAX_METHOD == "minimize":
        magnet.minimize(tol=MINIMIZE_TOL, nsamples=MINIMIZE_NSAMPLES)
    elif RELAX_METHOD == "relax":
        magnet.relax(tol=RELAX_TOL)
    else:
        raise ValueError("RELAX_METHOD must be 'minimize' or 'relax'.")

    print("Final average m:", magnet.magnetization.average())

    if SAVE_FINAL_OVF:
        magnet.magnetization.save_ovf(
            os.path.join(case_ovf_dir, f"{case_name}_relaxed_magnetization.ovf")
        )

    m_final = get_m_array(magnet)

    Q_final = estimate_topological_charge(m_final, geometry2d, dx, dy)

    area_final, radius_final = estimate_reversed_core_area_radius(
        m_final,
        geometry2d,
        SK_CORE_POLARITY,
    )

    print(f"{case_name}: final approximate Q = {Q_final:+.3f}")
    print(f"{case_name}: final reversed-core radius = {radius_final*1e9:.2f} nm")

    plot_case_results(case_name, m_final)

    mavg = magnet.magnetization.average()

    summary_rows.append({
        "case": case_name,
        "Nx": Nx,
        "Ny": Ny,
        "Nz": Nz,
        "dx_m": dx,
        "dy_m": dy,
        "dz_m": dz,
        "grain_size_m": grain_size if case_name != "no_grains" else 0.0,
        "n_active_grains": len(active_region_ids) if case_name != "no_grains" else 0,
        "Msat_A_per_m": MS0,
        "Aex_J_per_m": AEX0,
        "DMI_J_per_m2": D0,
        "Ku1_J_per_m3": KU10,
        "alpha": ALPHA,
        "bias_Bz_T": BIAS_BZ_T,
        "KU1_SIGMA_applied": KU1_SIGMA if case_name in ["ku1_only", "all_three"] else 0.0,
        "MS_SIGMA_applied": MS_SIGMA if case_name in ["msat_only", "all_three"] else 0.0,
        "D_SIGMA_applied": D_SIGMA if case_name in ["dmi_only", "all_three"] else 0.0,
        "correlate_ku_dmi": CORRELATE_KU_DMI,
        "skyrmion_seed_radius_m": SK_RADIUS,
        "skyrmion_seed_wall_width_m": SK_WALL_WIDTH,
        "skyrmion_core_polarity": SK_CORE_POLARITY,
        "neel_chirality": NEEL_CHIRALITY,
        "Q_initial_approx": Q_initial,
        "Q_final_approx": Q_final,
        "initial_core_area_m2": area_initial,
        "final_core_area_m2": area_final,
        "initial_core_radius_m": radius_initial,
        "final_core_radius_m": radius_final,
        "mavg_x_final": mavg[0],
        "mavg_y_final": mavg[1],
        "mavg_z_final": mavg[2],
    })

    # Save a running summary immediately after this variation completes.
    running_summary_df = pd.DataFrame(summary_rows)
    running_summary_path = os.path.join(csv_dir, "running_skyrmion_relax_summary_by_case.csv")
    running_summary_df.to_csv(running_summary_path, index=False)
    print("Saved running summary:", running_summary_path)

    # Mark this variation as completed.
    done_marker = os.path.join(case_ovf_dir, f"{case_name}_COMPLETED.txt")
    with open(done_marker, "w") as f:
        f.write(f"Completed case: {case_name}\n")
        f.write(f"Q_final_approx: {Q_final:+.6f}\n")
        f.write(f"final_core_radius_nm: {radius_final*1e9:.6f}\n")
        f.write(f"mavg_z_final: {mavg[2]:+.6f}\n")

    # Zip and download this completed variation before moving on.
    zip_and_download_finished_case(case_name)

    del magnet
    gc.collect()

# =========================================================
# 17) Save summary and comparison plots
# =========================================================
summary_df = pd.DataFrame(summary_rows)

summary_path = os.path.join(csv_dir, "skyrmion_relax_summary_by_case.csv")
summary_df.to_csv(summary_path, index=False)
print("Saved summary:", summary_path)

display(summary_df)

# Final Q comparison.
plt.figure(figsize=(7, 5))
plt.bar(summary_df["case"], summary_df["Q_final_approx"])
plt.axhline(+1, linestyle="--", linewidth=1)
plt.axhline(-1, linestyle="--", linewidth=1)
plt.axhline(0, linestyle="-", linewidth=1)
plt.xticks(rotation=30, ha="right")
plt.ylabel("Approximate final Q")
plt.title("Final skyrmion topological charge by disorder case")
plt.tight_layout()

path = os.path.join(plot_dir, "comparison_final_Q.png")
plt.savefig(path, dpi=200, bbox_inches="tight")
plt.show()
print("Saved:", path)

# Final core radius comparison.
plt.figure(figsize=(7, 5))
plt.bar(summary_df["case"], summary_df["final_core_radius_m"] * 1e9)
plt.xticks(rotation=30, ha="right")
plt.ylabel("Final reversed-core radius (nm)")
plt.title("Final skyrmion core radius by disorder case")
plt.tight_layout()

path = os.path.join(plot_dir, "comparison_final_core_radius.png")
plt.savefig(path, dpi=200, bbox_inches="tight")
plt.show()
print("Saved:", path)

# Final average mz comparison.
plt.figure(figsize=(7, 5))
plt.bar(summary_df["case"], summary_df["mavg_z_final"])
plt.xticks(rotation=30, ha="right")
plt.ylabel("Final <mz>")
plt.title("Final average mz by disorder case")
plt.tight_layout()

path = os.path.join(plot_dir, "comparison_final_mz_avg.png")
plt.savefig(path, dpi=200, bbox_inches="tight")
plt.show()
print("Saved:", path)

# =========================================================
# 18) Zip all outputs
# =========================================================
zip_name = "cofeb_grain_skyrmion_relax_outputs.zip"

with zipfile.ZipFile(zip_name, "w") as z:
    for root, dirs, files_here in os.walk(base_out_dir):
        for f in files_here:
            if f.endswith((".png", ".csv", ".ovf")):
                path = os.path.join(root, f)
                z.write(path)

print("Created:", zip_name)

# Uncomment in Colab for automatic download:
# from google.colab import files
# files.download(zip_name)

print("Done.")
```

```python
# =========================================================
# MINIMAL BOOTSTRAP ONLY
# Recreates geometry + Ku1 grain disorder needed for depinning.
# Does NOT relax.
# Does NOT run no_grains/ku1_only loop.
# Does NOT save material OVFs.
# =========================================================

import numpy as np
import os
from PIL import Image

from mumaxplus import Grid, World
from mumaxplus.util import VoronoiTessellator

# =========================================================
# 0) Files
# =========================================================

# Same mask used for the relaxation run.
shape_path = "/content/full_mask.png"

# Same relaxed OVF you want to drive.
RELAXED_OVF_PATH = "/content/no_grains_relaxed_magnetization.ovf"

if not os.path.exists(shape_path):
    raise FileNotFoundError(
        f"Could not find mask: {shape_path}\n"
        "Upload full_mask.png to /content or change shape_path."
    )

if not os.path.exists(RELAXED_OVF_PATH):
    raise FileNotFoundError(
        f"Could not find relaxed OVF: {RELAXED_OVF_PATH}\n"
        "Upload the relaxed ku1_only OVF to /content or change RELAXED_OVF_PATH."
    )

print("Using mask:", shape_path)
print("Using relaxed OVF:", RELAXED_OVF_PATH)

# =========================================================
# 1) Grid and cell size
# MUST MATCH the relaxation run exactly
# =========================================================

USE_IMAGE_SIZE = True

if USE_IMAGE_SIZE:
    img_tmp = Image.open(shape_path).convert("L")
    Nx, Ny = img_tmp.size
    img_tmp.close()
else:
    Nx = 512
    Ny = 512

Nz = 10

dx = 2e-9
dy = 2e-9
dz = 1.0e-10

print(f"Grid: Nx={Nx}, Ny={Ny}, Nz={Nz}")
print(f"Cell size: dx={dx*1e9:.3f} nm, dy={dy*1e9:.3f} nm, dz={dz*1e9:.3f} nm")

# =========================================================
# 2) Material values
# MUST MATCH the relaxation run exactly
# =========================================================

MS0 = 1100e3
AEX0 = 15e-12
D0 = 1.2e-3
KU10 = 1.2e6

ALPHA = 0.5
BIAS_BZ_T = 40e-3

# =========================================================
# 3) Grain/disorder settings
# MUST MATCH the relaxation run exactly
# =========================================================

grain_size = 25e-9

# IMPORTANT:
# Change this to the reduced value you actually used for the successful ku1_only run.
MAX_REGION_ID = 1000

KU1_SIGMA = 0.05
MS_SIGMA  = 0.05
D_SIGMA   = 0.05

CORRELATE_KU_DMI = False
RANDOM_SEED = 12345

INVERT_MASK = False

# Skyrmion polarity from your relaxation script.
SK_CORE_POLARITY = -1
CORE_POLARITY = SK_CORE_POLARITY

# =========================================================
# 4) Output folders
# =========================================================

base_out_dir = "cofeb_grain_skyrmion_relax_outputs"

plot_dir = os.path.join(base_out_dir, "plots")
ovf_dir = os.path.join(base_out_dir, "ovf")
csv_dir = os.path.join(base_out_dir, "csv")
grain_dir = os.path.join(base_out_dir, "grain_maps")

for d in [base_out_dir, plot_dir, ovf_dir, csv_dir, grain_dir]:
    os.makedirs(d, exist_ok=True)

# =========================================================
# 5) Recreate geometry from mask
# =========================================================

def load_geometry_from_image(path, nx, ny, nz, threshold=128, invert=False):
    img = Image.open(path).convert("L")
    img = img.resize((nx, ny), Image.Resampling.NEAREST)
    arr = np.array(img)

    if invert:
        mask2d = arr > threshold
    else:
        mask2d = arr < threshold

    geometry = np.repeat(mask2d[np.newaxis, :, :], nz, axis=0).astype(bool)

    magnetic_fraction = geometry.mean()
    print(f"Magnetic fraction of grid: {magnetic_fraction:.4f}")

    if magnetic_fraction < 0.001:
        print("WARNING: geometry is almost empty. Try INVERT_MASK = True.")
    if magnetic_fraction > 0.999:
        print("WARNING: geometry is almost full. Check whether the mask is inverted.")

    return geometry

geometry = load_geometry_from_image(
    shape_path,
    Nx,
    Ny,
    Nz,
    invert=INVERT_MASK,
)

geometry2d = geometry[0]

# =========================================================
# 6) Recreate the same Voronoi grain map
# =========================================================

region_world = World(cellsize=(dx, dy, dz))
region_grid = Grid((Nx, Ny, Nz))

tessellator = VoronoiTessellator(
    grain_size,
    seed=RANDOM_SEED,
    max_idx=MAX_REGION_ID,
)

regions = tessellator.generate(region_world, region_grid, make_2d=True)
regions = np.asarray(regions).astype(np.int32)

if regions.ndim == 2:
    regions = regions[np.newaxis, :, :]

if regions.shape[0] == 1 and Nz > 1:
    regions = np.repeat(regions, Nz, axis=0)

if regions.shape != geometry.shape:
    raise ValueError(
        f"regions shape {regions.shape} does not match geometry shape {geometry.shape}"
    )

regions2d = regions[0]

active_region_ids = np.unique(regions[geometry])
active_region_ids = [int(i) for i in active_region_ids]

if len(active_region_ids) == 0:
    raise RuntimeError("No active regions found inside magnetic geometry.")

print("Active grains inside magnetic geometry:", len(active_region_ids))
print("Lowest active region ID:", min(active_region_ids))
print("Highest active region ID:", max(active_region_ids))

# =========================================================
# 7) Recreate the same Ku1 disorder values
# =========================================================

rng = np.random.default_rng(RANDOM_SEED)

ku1_by_region = {}
msat_by_region = {}
dmi_by_region = {}

for rid in active_region_ids:
    if CORRELATE_KU_DMI:
        eta_interface = rng.normal()
        eta_ms = rng.normal()

        ku1_by_region[rid] = max(0.0, KU10 * (1.0 + KU1_SIGMA * eta_interface))
        dmi_by_region[rid] = max(0.0, D0   * (1.0 + D_SIGMA   * eta_interface))
        msat_by_region[rid] = max(1.0, MS0 * (1.0 + MS_SIGMA  * eta_ms))
    else:
        ku1_by_region[rid] = max(0.0, KU10 * (1.0 + KU1_SIGMA * rng.normal()))
        msat_by_region[rid] = max(1.0, MS0 * (1.0 + MS_SIGMA  * rng.normal()))
        dmi_by_region[rid] = max(0.0, D0   * (1.0 + D_SIGMA   * rng.normal()))

print("Minimal bootstrap complete.")
print("You can now run the OVF depinning cell.")
```

```python
# =========================================================
# NEW CELL ONLY:
# OVF-start 3 x 10 ns SOT staircase depinning search
#
# Assumes previous setup cell has already been run, so these exist:
#   geometry, geometry2d, regions, active_region_ids, ku1_by_region
#   Nx, Ny, Nz, dx, dy, dz
#   MS0, AEX0, D0, KU10, ALPHA, BIAS_BZ_T
#   SK_CORE_POLARITY or SK_POLARIZATION
#   grain_size, RANDOM_SEED
#
# This cell DOES NOT rerun mask import, grain generation, or relaxation.
# =========================================================

import os
import gc
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    from google.colab import files
    IN_COLAB = True
except Exception:
    IN_COLAB = False

from mumaxplus import Ferromagnet, Grid, World

# =========================================================
# 0) Required existing variables check
# =========================================================

required_names = [
    "geometry",
    "geometry2d",
    "regions",
    "active_region_ids",
    "ku1_by_region",
    "Nx",
    "Ny",
    "Nz",
    "dx",
    "dy",
    "dz",
    "MS0",
    "AEX0",
    "D0",
    "KU10",
    "ALPHA",
    "BIAS_BZ_T",
    "grain_size",
    "RANDOM_SEED",
]

missing = [name for name in required_names if name not in globals()]

if missing:
    raise RuntimeError(
        "This new cell needs the earlier setup cell to have already run.\n"
        "Missing variables:\n"
        + "\n".join(missing)
        + "\n\nIf the runtime restarted, rerun the setup cell first."
    )

if "base_out_dir" not in globals():
    base_out_dir = "cofeb_grain_skyrmion_relax_outputs"

if "plot_dir" not in globals():
    plot_dir = os.path.join(base_out_dir, "plots")

if "csv_dir" not in globals():
    csv_dir = os.path.join(base_out_dir, "csv")

if "ovf_dir" not in globals():
    ovf_dir = os.path.join(base_out_dir, "ovf")

for d in [base_out_dir, plot_dir, csv_dir, ovf_dir]:
    os.makedirs(d, exist_ok=True)

# Works with either your manual-seed script or the later API-seed naming.
CORE_POLARITY = globals().get("SK_CORE_POLARITY", globals().get("SK_POLARIZATION", -1))

print("Existing setup found.")
print(f"Grid: {Nx} x {Ny} x {Nz}")
print(f"Cell size: dx={dx*1e9:.3f} nm, dy={dy*1e9:.3f} nm, dz={dz*1e9:.3f} nm")
print(f"Using core polarity: {CORE_POLARITY}")

# =========================================================
# 1) User controls
# =========================================================

# Set this to the relaxed OVF already uploaded to /content.
# Example:
# RELAXED_OVF_PATH = "/content/ku1_only_relaxed_magnetization.ovf"
RELAXED_OVF_PATH = "/content/no_grains_relaxed_magnetization.ovf"

DEPIN_CASE = "ku1_only"

# Current staircase:
# 1e11, 2e11, 3e11, ...
J_START = 1e11
J_STEP = 1e11
J_MAX = 2.0e12

# Pulse protocol:
# each current gets up to 3 pulses, each 10 ns.
N_PULSES_PER_J = 3
PULSE_TIME = 10e-9

# Let it settle after each pulse.
# Set 0.0 for back-to-back pulses.
INTER_PULSE_SETTLE_TIME = 2e-9

# Depinning criterion:
# moved by at least one grain length.
DEPIN_DISTANCE_NM = grain_size * 1e9

# Survival criteria.
MIN_FINAL_RADIUS_NM = 10.0
MIN_FINAL_Q_ABS = 0.5

# Clean threshold mode:
# True = every current starts from the same relaxed OVF.
# This is best for finding J_dep.
RESET_EACH_CURRENT_FROM_OVF = True

# Refine after first depinning current is found.
REFINE_AFTER_FIRST_DEPIN = True
N_REFINE_STEPS = 5

# =========================================================
# 2) Pt/CoFeB/Ir SOT settings
# =========================================================
# This is NOT Zhang-Li.
# It uses Slonczewski/SOT-like drive.

SOT_POL = 0.10
SOT_LAMBDA = 1.0
SOT_EPSILON_PRIME = 0.0
FREE_LAYER_THICKNESS = Nz * dz

# Effective perpendicular spin-current injection for SOT form.
SOT_CURRENT_DIRECTION = (0.0, 0.0, 1.0)

# For charge current along +x in Pt/Ir, spin Hall polarization is usually +/- y.
# Flip to (0.0, -1.0, 0.0) if motion is opposite to expected.
SPIN_POLARIZATION_DIRECTION = (0.0, 1.0, 0.0)

# Flip if the Pt/Ir stack sign is opposite.
FIXED_LAYER_ON_TOP = False

depin_base_dir = os.path.join(base_out_dir, "ovf_3pulse_10ns_depinning")
os.makedirs(depin_base_dir, exist_ok=True)

# =========================================================
# 3) OVF loader
# =========================================================

def read_ovf_vector_field_to_numpy(ovf_path):
    """
    Fallback OVF reader for vector-field OVFs.
    Returns array shaped (3, Nz, Ny, Nx).

    Handles OVF text, binary 4, and binary 8 in the common mumax/OVF layout.
    """

    if not os.path.exists(ovf_path):
        raise FileNotFoundError(f"Cannot find OVF: {ovf_path}")

    header = []
    mode = None

    with open(ovf_path, "rb") as f:
        while True:
            line_b = f.readline()
            if not line_b:
                raise RuntimeError("Reached end of file before OVF data block.")

            line = line_b.decode("utf-8", errors="ignore").strip()
            header.append(line)

            low = line.lower()

            if "begin: data text" in low:
                mode = "text"
                break
            elif "begin: data binary 4" in low:
                mode = "binary4"
                break
            elif "begin: data binary 8" in low:
                mode = "binary8"
                break

        xnodes = ynodes = znodes = None

        for h in header:
            h_clean = h.replace("#", "").strip()
            parts = h_clean.split(":", 1)

            if len(parts) != 2:
                continue

            key = parts[0].strip().lower()
            val = parts[1].strip()

            if key == "xnodes":
                xnodes = int(float(val))
            elif key == "ynodes":
                ynodes = int(float(val))
            elif key == "znodes":
                znodes = int(float(val))

        if xnodes is None or ynodes is None or znodes is None:
            raise RuntimeError("Could not read xnodes/ynodes/znodes from OVF header.")

        if (xnodes, ynodes, znodes) != (Nx, Ny, Nz):
            raise ValueError(
                "OVF grid does not match current notebook grid.\n"
                f"OVF:     Nx={xnodes}, Ny={ynodes}, Nz={znodes}\n"
                f"Notebook Nx={Nx}, Ny={Ny}, Nz={Nz}"
            )

        n_vec = xnodes * ynodes * znodes
        n_vals = 3 * n_vec

        if mode == "text":
            text = f.read().decode("utf-8", errors="ignore")
            tokens = []

            for line in text.splitlines():
                s = line.strip()
                if not s or s.startswith("#"):
                    continue
                tokens.extend(s.split())

            data = np.asarray(tokens[:n_vals], dtype=np.float64)

        elif mode == "binary4":
            raw = np.fromfile(f, dtype=np.float32)

            # First binary value is usually the OVF check number.
            if raw.size >= n_vals + 1:
                raw = raw[1:1+n_vals]

            data = raw.astype(np.float64)

        elif mode == "binary8":
            raw = np.fromfile(f, dtype=np.float64)

            # First binary value is usually the OVF check number.
            if raw.size >= n_vals + 1:
                raw = raw[1:1+n_vals]

            data = raw

        else:
            raise RuntimeError(f"Unsupported OVF data mode: {mode}")

    if data.size < n_vals:
        raise RuntimeError(
            f"OVF data too short. Got {data.size} values, expected {n_vals}."
        )

    data = data[:n_vals]

    # OVF ordering: x fastest, then y, then z, each point has mx,my,mz.
    arr = data.reshape((Nz, Ny, Nx, 3))
    m = np.moveaxis(arr, -1, 0).astype(np.float32)

    return m


def load_relaxed_ovf_into_magnet(magnet, ovf_path):
    """
    Load relaxed OVF into magnetization.

    First tries mumax+'s native load_ovf if available.
    If unavailable, reads the OVF into a NumPy array and assigns it.
    """

    if not os.path.exists(ovf_path):
        raise FileNotFoundError(
            f"RELAXED_OVF_PATH does not exist:\n{ovf_path}\n"
            "Upload your relaxed OVF to /content and set RELAXED_OVF_PATH."
        )

    if hasattr(magnet.magnetization, "load_ovf"):
        try:
            magnet.magnetization.load_ovf(ovf_path)
            print("Loaded relaxed OVF using native load_ovf:", ovf_path)
            return
        except Exception as e:
            print("Native load_ovf failed; trying fallback OVF reader.")
            print("Native load_ovf error:", repr(e))

    m_loaded = read_ovf_vector_field_to_numpy(ovf_path)
    magnet.magnetization = m_loaded
    print("Loaded relaxed OVF using fallback NumPy reader:", ovf_path)

# =========================================================
# 4) Analysis helpers
# =========================================================

def get_m_array_local(magnet):
    m = np.asarray(magnet.magnetization.eval())

    expected_shape = (3, Nz, Ny, Nx)
    if m.shape != expected_shape:
        raise ValueError(
            f"Unexpected magnetization shape: got {m.shape}, expected {expected_shape}"
        )

    return m


def estimate_topological_charge_local(m_array):
    m = m_array[:, 0, :, :]

    dmdy, dmdx = np.gradient(m, dy, dx, axis=(1, 2))

    m_vec = np.moveaxis(m, 0, -1)
    dmdx_vec = np.moveaxis(dmdx, 0, -1)
    dmdy_vec = np.moveaxis(dmdy, 0, -1)

    cross = np.cross(dmdx_vec, dmdy_vec)
    q_density = np.einsum("ijk,ijk->ij", m_vec, cross)
    q_density = np.where(geometry2d, q_density, 0.0)

    Q = np.sum(q_density) * dx * dy / (4 * np.pi)

    return float(Q)


def estimate_reversed_core_area_radius_local(m_array):
    mz = m_array[2, 0, :, :]

    if CORE_POLARITY == -1:
        core_mask = (mz < 0.0) & geometry2d
    else:
        core_mask = (mz > 0.0) & geometry2d

    area = np.sum(core_mask) * dx * dy
    radius = np.sqrt(area / np.pi) if area > 0 else 0.0

    return float(area), float(radius)


def weighted_core_position(m_array):
    """
    Weighted skyrmion core center.

    For CORE_POLARITY = -1, negative mz is the core.
    For CORE_POLARITY = +1, positive mz is the core.
    """

    mz = m_array[2, 0, :, :]

    core_weight = np.clip(CORE_POLARITY * mz, 0.0, 1.0)
    core_weight = np.where(geometry2d, core_weight, 0.0)

    total_weight = np.sum(core_weight)

    if total_weight <= 0:
        return np.nan, np.nan, 0.0, False

    yy, xx = np.indices(mz.shape)

    x_m = np.sum(core_weight * xx) / total_weight * dx
    y_m = np.sum(core_weight * yy) / total_weight * dy

    if CORE_POLARITY == -1:
        core_mask = (mz < 0.0) & geometry2d
    else:
        core_mask = (mz > 0.0) & geometry2d

    area = np.sum(core_mask) * dx * dy
    radius_m = np.sqrt(area / np.pi) if area > 0 else 0.0

    return float(x_m), float(y_m), float(radius_m), True

# =========================================================
# 5) Build magnet from relaxed OVF
# =========================================================

def build_ku1_depinning_magnet_from_ovf(ovf_path, name="ku1_depinning"):
    """
    Rebuild only the ku1_only magnet using existing geometry/grain data,
    then load relaxed OVF as the starting magnetization.
    """

    case_world = World(cellsize=(dx, dy, dz))
    case_grid = Grid((Nx, Ny, Nz))

    magnet = Ferromagnet(
        case_world,
        case_grid,
        geometry=geometry,
        regions=regions,
        name=name,
    )

    magnet.alpha = ALPHA
    magnet.aex = AEX0
    magnet.msat = MS0
    magnet.ku1 = KU10
    magnet.anisU = (0, 0, 1)
    magnet.dmi_tensor.set_interfacial_dmi(D0)

    magnet.bias_magnetic_field = (0, 0, BIAS_BZ_T)

    # Start with torques off.
    magnet.enable_zhang_li_torque = False
    magnet.enable_slonczewski_torque = False

    # Apply only Ku1 disorder.
    for rid in active_region_ids:
        magnet.ku1.set_in_region(rid, ku1_by_region[rid])

    load_relaxed_ovf_into_magnet(magnet, ovf_path)

    return case_world, magnet

# =========================================================
# 6) SOT drive helpers
# =========================================================

def enable_sot_drive(magnet, J):
    """
    SOT-like drive using Slonczewski torque.
    This is not Zhang-Li.
    """

    magnet.enable_zhang_li_torque = False
    magnet.enable_slonczewski_torque = True

    magnet.jcur = (
        J * SOT_CURRENT_DIRECTION[0],
        J * SOT_CURRENT_DIRECTION[1],
        J * SOT_CURRENT_DIRECTION[2],
    )

    magnet.pol = SOT_POL
    magnet.fixed_layer = SPIN_POLARIZATION_DIRECTION
    magnet.Lambda = SOT_LAMBDA
    magnet.epsilon_prime = SOT_EPSILON_PRIME
    magnet.free_layer_thickness = FREE_LAYER_THICKNESS
    magnet.fixed_layer_on_top = FIXED_LAYER_ON_TOP


def disable_sot_drive(magnet):
    magnet.jcur = (0.0, 0.0, 0.0)
    magnet.enable_slonczewski_torque = False
    magnet.enable_zhang_li_torque = False

# =========================================================
# 7) Zip/download helper
# =========================================================

def zip_and_download_folder(folder, zip_name):
    zip_path = os.path.join(base_out_dir, zip_name)

    with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as z:
        for root, dirs, files_here in os.walk(folder):
            for f in files_here:
                if f.endswith((".png", ".csv", ".ovf", ".txt", ".npy")):
                    path = os.path.join(root, f)
                    arcname = os.path.relpath(path, base_out_dir)
                    z.write(path, arcname)

    print("Created:", zip_path)

    if IN_COLAB:
        files.download(zip_path)

    return zip_path

# =========================================================
# 8) One current trial: 3 pulses x 10 ns
# =========================================================

def run_three_pulse_trial_from_ovf(J, ovf_path, trial_label):
    safe_J_label = f"{J:.3e}".replace("+", "").replace(".", "p")
    trial_name = f"{trial_label}_J_{safe_J_label}"

    trial_dir = os.path.join(depin_base_dir, trial_name)
    trial_plot_dir = os.path.join(trial_dir, "plots")
    trial_csv_dir = os.path.join(trial_dir, "csv")
    trial_ovf_dir = os.path.join(trial_dir, "ovf")

    for d in [trial_dir, trial_plot_dir, trial_csv_dir, trial_ovf_dir]:
        os.makedirs(d, exist_ok=True)

    print("\n" + "=" * 72)
    print(f"OVF-start depinning trial: J = {J:.3e} A/m²")
    print("=" * 72)

    world, magnet = build_ku1_depinning_magnet_from_ovf(
        ovf_path=ovf_path,
        name=trial_name,
    )

    m_initial = get_m_array_local(magnet)

    x0, y0, r0, alive0 = weighted_core_position(m_initial)

    if not alive0:
        status = "DESTROYED_AT_START"

        result = {
            "trial_name": trial_name,
            "J_A_per_m2": J,
            "status": status,
            "pulse_count": 0,
            "final_displacement_nm": np.nan,
            "final_radius_nm": 0.0,
            "Q_final_approx": np.nan,
            "source_ovf": ovf_path,
        }

        pd.DataFrame([result]).to_csv(
            os.path.join(trial_csv_dir, f"{trial_name}_result.csv"),
            index=False,
        )

        print("No skyrmion core found in starting OVF.")
        return result

    print(f"Initial core: x={x0*1e9:.2f} nm, y={y0*1e9:.2f} nm, radius={r0*1e9:.2f} nm")

    rows = []
    status = "PINNED_AFTER_3"

    for pulse_idx in range(1, N_PULSES_PER_J + 1):
        print(f"Pulse {pulse_idx}/{N_PULSES_PER_J} at J = {J:.3e} A/m²")

        # Pulse on.
        enable_sot_drive(magnet, J)
        world.timesolver.run(PULSE_TIME)

        # Pulse off.
        disable_sot_drive(magnet)

        # Optional settle.
        if INTER_PULSE_SETTLE_TIME > 0:
            world.timesolver.run(INTER_PULSE_SETTLE_TIME)

        m_now = get_m_array_local(magnet)

        x, y, radius_m, alive = weighted_core_position(m_now)

        if alive:
            displacement_nm = np.sqrt((x - x0)**2 + (y - y0)**2) * 1e9
            radius_nm = radius_m * 1e9
        else:
            displacement_nm = np.nan
            radius_nm = 0.0

        row = {
            "trial_name": trial_name,
            "J_A_per_m2": J,
            "pulse_index": pulse_idx,
            "pulse_time_ns": PULSE_TIME * 1e9,
            "settle_time_ns": INTER_PULSE_SETTLE_TIME * 1e9,
            "total_elapsed_ns": pulse_idx * (PULSE_TIME + INTER_PULSE_SETTLE_TIME) * 1e9,
            "core_x_nm": x * 1e9 if alive else np.nan,
            "core_y_nm": y * 1e9 if alive else np.nan,
            "core_radius_nm": radius_nm,
            "displacement_nm": displacement_nm,
            "alive": alive,
        }

        rows.append(row)

        traj_df = pd.DataFrame(rows)
        traj_path = os.path.join(trial_csv_dir, f"{trial_name}_pulse_trajectory.csv")
        traj_df.to_csv(traj_path, index=False)

        print(f"  displacement = {displacement_nm:.2f} nm")
        print(f"  radius       = {radius_nm:.2f} nm")

        if not alive:
            status = "DESTROYED"
            print("Skyrmion destroyed.")
            break

        if displacement_nm >= DEPIN_DISTANCE_NM:
            status = "DEPINNED"
            print(f"Depinned after pulse {pulse_idx}.")
            break

    m_final = get_m_array_local(magnet)

    Q_final = estimate_topological_charge_local(m_final)

    area_final, radius_final = estimate_reversed_core_area_radius_local(m_final)
    final_radius_nm = radius_final * 1e9

    if len(rows) > 0:
        final_displacement_nm = rows[-1]["displacement_nm"]
        pulse_count = rows[-1]["pulse_index"]
    else:
        final_displacement_nm = np.nan
        pulse_count = 0

    survived = (
        np.isfinite(Q_final)
        and abs(Q_final) >= MIN_FINAL_Q_ABS
        and final_radius_nm >= MIN_FINAL_RADIUS_NM
    )

    if status == "DEPINNED" and not survived:
        status = "DESTROYED_AFTER_MOVING"

    result = {
        "trial_name": trial_name,
        "J_A_per_m2": J,
        "status": status,
        "pulse_count": pulse_count,
        "pulse_time_ns": PULSE_TIME * 1e9,
        "settle_time_ns": INTER_PULSE_SETTLE_TIME * 1e9,
        "depin_distance_nm": DEPIN_DISTANCE_NM,
        "final_displacement_nm": final_displacement_nm,
        "final_radius_nm": final_radius_nm,
        "Q_final_approx": Q_final,
        "survived": survived,
        "source_ovf": ovf_path,
    }

    result_path = os.path.join(trial_csv_dir, f"{trial_name}_result.csv")
    pd.DataFrame([result]).to_csv(result_path, index=False)
    print("Saved result:", result_path)

    final_ovf_path = os.path.join(trial_ovf_dir, f"{trial_name}_final_magnetization.ovf")
    magnet.magnetization.save_ovf(final_ovf_path)
    print("Saved final OVF:", final_ovf_path)

    traj_df = pd.DataFrame(rows)

    if len(traj_df) > 0:
        plt.figure(figsize=(6, 5))
        plt.plot(traj_df["pulse_index"], traj_df["displacement_nm"], marker="o")
        plt.axhline(DEPIN_DISTANCE_NM, linestyle="--", linewidth=1)
        plt.xlabel("pulse number")
        plt.ylabel("displacement after pulse (nm)")
        plt.title(f"{trial_name}: {status}")
        plt.tight_layout()

        disp_plot_path = os.path.join(trial_plot_dir, f"{trial_name}_pulse_displacement.png")
        plt.savefig(disp_plot_path, dpi=200, bbox_inches="tight")
        plt.show()
        print("Saved:", disp_plot_path)

        plt.figure(figsize=(6, 5))
        plt.plot(traj_df["core_x_nm"], traj_df["core_y_nm"], marker="o")
        plt.xlabel("core x (nm)")
        plt.ylabel("core y (nm)")
        plt.title(f"{trial_name}: trajectory, {status}")
        plt.axis("equal")
        plt.tight_layout()

        traj_plot_path = os.path.join(trial_plot_dir, f"{trial_name}_trajectory.png")
        plt.savefig(traj_plot_path, dpi=200, bbox_inches="tight")
        plt.show()
        print("Saved:", traj_plot_path)

    marker_path = os.path.join(trial_dir, f"{trial_name}_{status}.txt")
    with open(marker_path, "w") as f:
        for k, v in result.items():
            f.write(f"{k}: {v}\n")

    zip_and_download_folder(
        trial_dir,
        f"{trial_name}_outputs.zip",
    )

    del magnet
    del world
    gc.collect()

    return result

# =========================================================
# 9) Staircase search and optional refinement
# =========================================================

def ovf_staircase_depinning_search(ovf_path, trial_label):
    results = []

    last_pinned_J = None
    first_depinned_J = None

    J_values = np.arange(
        J_START,
        J_MAX + 0.5 * J_STEP,
        J_STEP,
    )

    for J in J_values:
        result = run_three_pulse_trial_from_ovf(
            J=J,
            ovf_path=ovf_path,
            trial_label=trial_label,
        )

        results.append(result)

        running_df = pd.DataFrame(results)
        running_path = os.path.join(
            depin_base_dir,
            f"{trial_label}_staircase_running_results.csv",
        )
        running_df.to_csv(running_path, index=False)
        print("Saved running staircase results:", running_path)

        if result["status"] == "DEPINNED":
            first_depinned_J = J
            print(f"First depinning current found: {J:.3e} A/m²")
            break

        elif result["status"] == "PINNED_AFTER_3":
            last_pinned_J = J

        elif result["status"].startswith("DESTROYED"):
            print("Skyrmion destroyed before clean depinning.")
            print("Try smaller J_STEP/J_MAX, or flip SPIN_POLARIZATION_DIRECTION.")
            break

    final_df = pd.DataFrame(results)

    final_path = os.path.join(
        depin_base_dir,
        f"{trial_label}_staircase_final_results.csv",
    )
    final_df.to_csv(final_path, index=False)

    print("Saved final staircase results:", final_path)

    return final_df, last_pinned_J, first_depinned_J


def refine_ovf_depinning_bracket(ovf_path, low_J, high_J, trial_label):
    if low_J is None or high_J is None:
        print("No valid bracket for refinement.")
        return None

    print("\n" + "=" * 72)
    print("Refining depinning bracket")
    print(f"Initial bracket: {low_J:.3e} < J_dep <= {high_J:.3e}")
    print("=" * 72)

    refine_results = []

    low = low_J
    high = high_J

    for step in range(1, N_REFINE_STEPS + 1):
        J_mid = 0.5 * (low + high)

        result = run_three_pulse_trial_from_ovf(
            J=J_mid,
            ovf_path=ovf_path,
            trial_label=f"{trial_label}_refine_{step}",
        )

        refine_results.append(result)

        if result["status"] == "DEPINNED":
            high = J_mid
        elif result["status"] == "PINNED_AFTER_3":
            low = J_mid
        elif result["status"].startswith("DESTROYED"):
            high = J_mid

        print(f"Updated bracket: {low:.3e} < J_dep <= {high:.3e}")

        refine_df = pd.DataFrame(refine_results)
        refine_path = os.path.join(
            depin_base_dir,
            f"{trial_label}_refinement_running_results.csv",
        )
        refine_df.to_csv(refine_path, index=False)

    print("\nRefined depinning estimate:")
    print(f"J_dep ≈ {high:.3e} A/m²")
    print(f"Protocol: {N_PULSES_PER_J} pulses × {PULSE_TIME*1e9:.1f} ns")

    return high

# =========================================================
# 10) Run it
# =========================================================

if not os.path.exists(RELAXED_OVF_PATH):
    raise FileNotFoundError(
        f"RELAXED_OVF_PATH does not exist:\n{RELAXED_OVF_PATH}\n"
        "Upload the relaxed OVF to /content and update the path."
    )

trial_label = f"seed_{RANDOM_SEED}_{DEPIN_CASE}_OVF_3pulse_10ns"

staircase_df, last_pinned_J, first_depinned_J = ovf_staircase_depinning_search(
    ovf_path=RELAXED_OVF_PATH,
    trial_label=trial_label,
)

J_dep_refined = None

if REFINE_AFTER_FIRST_DEPIN and last_pinned_J is not None and first_depinned_J is not None:
    J_dep_refined = refine_ovf_depinning_bracket(
        ovf_path=RELAXED_OVF_PATH,
        low_J=last_pinned_J,
        high_J=first_depinned_J,
        trial_label=trial_label,
    )

summary = {
    "trial_label": trial_label,
    "source_ovf": RELAXED_OVF_PATH,
    "protocol": f"{N_PULSES_PER_J} pulses x {PULSE_TIME*1e9:.1f} ns",
    "inter_pulse_settle_ns": INTER_PULSE_SETTLE_TIME * 1e9,
    "depin_distance_nm": DEPIN_DISTANCE_NM,
    "last_pinned_J_A_per_m2": last_pinned_J,
    "first_depinned_J_A_per_m2": first_depinned_J,
    "refined_J_dep_A_per_m2": J_dep_refined,
    "SOT_POL": SOT_POL,
    "SPIN_POLARIZATION_DIRECTION": SPIN_POLARIZATION_DIRECTION,
    "FIXED_LAYER_ON_TOP": FIXED_LAYER_ON_TOP,
}

summary_path = os.path.join(
    depin_base_dir,
    f"{trial_label}_depinning_summary.csv",
)

pd.DataFrame([summary]).to_csv(summary_path, index=False)
print("Saved depinning summary:", summary_path)

zip_and_download_folder(
    depin_base_dir,
    f"{trial_label}_ALL_depinning_outputs.zip",
)

print("Done.")
print("Last pinned J:", last_pinned_J)
print("First depinned J:", first_depinned_J)
print("Refined J_dep:", J_dep_refined)
```
